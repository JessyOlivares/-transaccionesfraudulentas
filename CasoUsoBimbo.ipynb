```python
#Eliminación de variables categoricas y fechas
df_final = df_Completo.drop(['Fecha_Transaccion', 'Hora_transaccion', 'Referencia', 'Operador','Fecha_Apertura','Cliente', 'Tipo_de_Instrumento', 'Cuenta', 'Sucursal_nombre'], axis = 1, inplace = False)
df_final.dropna(inplace = True)
df_final
```

```python
#Tratamiento de l variable categoríca T_transaccion
le = LabelEncoder()
df_final['T_transaccion'] = le.fit_transform(df_final['T_transaccion'])
# Mostrar la asignación de categorías a valores numéricos
print("Asignación de categorías a valores numéricos:")
for categoria, codigo in zip(le.classes_, range(len(le.classes_))):
    print(f"{categoria}: {codigo}")
```

```python
#Calculo del montode Transferencia promedio por cuenta
df_final['meanMontoTransferencia'] = df_final.groupby('Numero_de_cuenta')['M_transaccion'].transform('mean')
df_final
```

#### Rango Intercuartílico (IQR)
El IQR es una medida estadística que se puede emplear para identificar outliers. Los valores que están más allá de 1.5 veces el IQR por debajo del primer cuartil (Q1) o por encima del tercer cuartil (Q3) se consideran outliers.

```python
# Calcular Q1 (primer cuartil) y Q3 (tercer cuartil)
df_iqr = df_final
Q1 = df_iqr['M_transaccion'].quantile(0.25)
Q3 = df_iqr['M_transaccion'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
df_iqr['is_outlier'] = (df_iqr['M_transaccion'] < lower_bound) | (df_iqr['M_transaccion'] > upper_bound)
df_iqr
```

```python
#Countplot de outliers obtenidos mediante el cálculo de IQR
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='is_outlier',data=df_iqr)

# Agregar anotaciones de conteo en cada barra
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                xy=(p.get_x() + p.get_width() / 2., height), 
                xytext=(0, 5),  # Desplazamiento de texto
                textcoords='offset points', 
                ha='center', va='center')

# Mostrar el gráfico
plt.title('Cantidad de ourliers mediante IQR')
plt.savefig('assets/ourliersIQR.png')
plt.show()


```

#### IsolationForest
Devulve la puntuación de anomalías de cada muestra mediante el algoritmo IsolationForest. 
El IsolationForest "aísla" las observaciones seleccionando aleatoriamente una entidad y, a continuación, seleccioar aleatoriamente un valor de división entre el máximo y el mínimo valores de la entidad seleccionada.

```python
df_IsolationForest = df_Multivariado.drop(['Numero_de_cuenta'], axis = 1, inplace = False)
```

```python
model = IsolationForest(random_state=42)
# Entrenar el modelo
model.fit(df_Multivariado.iloc[:, 1:])
```

```python
# Predecir outliers (-1) o inliers (1)
df_Multivariado['IsolationForest'] = model.predict(df_Multivariado.iloc[:, 1:])
df_Multivariado
```

```python
#Countplot de outliers obtenidos mediante el cálculo de IQR
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='IsolationForest',data=df_Multivariado)

# Agregar anotaciones de conteo en cada barra
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                xy=(p.get_x() + p.get_width() / 2., height), 
                xytext=(0, 5),  # Desplazamiento de texto
                textcoords='offset points', 
                ha='center', va='center')

# Mostrar el gráfico
plt.title('Cantidad de ourliers mediante IsolationForest')
plt.savefig('assets/ourliersIsolationForest.png')
plt.show()
```
